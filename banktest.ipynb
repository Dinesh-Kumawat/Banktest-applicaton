```python
last_account_num = 1002

```

```python
def admin():
    print("welcome to admin panel !!")
    admin = "admin"
    passwd = "admin@123"
    adm_usr = input("Enter the admin user name : ")
    if adm_usr == admin:
        password = getpass("Enter the Password : ")
        if password == passwd:
            print("!!All user details!!  ".center(120))
        
            for key,value in data.items():
                print("Account number : ",key)
                for k,v in value.items():
                    print(f"{k}={v}".rjust(100))
                print("\n")
        else:
            print("invalid password")
    else: 
        print("Invalid Admin User !!")
        admin()
        
        
            
```

```python
def credit(acc_no):
    bal = eval(input("\n\nEnter amount to deposit : "))
    old_bal = data[acc_no]['bal']
    data[acc_no]['bal'] += bal
    print("Your old balance was : ",old_bal)
    print(f"After creaditing {bal}, \nnow you have {data[acc_no]['bal']} in your account")
    print("Credit Sucessfully")

def debit(acc_no):
    bal = eval(input("\n\nEnter amount to withdraw : "))
    if bal > data[acc_no]['bal'] :
        print("\n\n Insufficient Balance in your account.... so enter the less amount")
        print(f"\nYou only have {data[acc_no]['bal']} rupees in your acount")
    else :
        data[acc_no]['bal'] -= bal
        print(f"\n\n{bal} rupees sucessfully debited from your account")
        print(f"\nYour updated balance is {data[acc_no]['bal']}")
        
        
def chk_bal(acc_no):
    print("\nYour account number is ",acc_no)
    for key,value in data[acc_no].items():
        if key == 'password' :
            continue
        print(f"{key} = {value}")
```

```python
def sub_menu(acc_no):
    menu=f"""
    {time.ctime()}
    Welcome back user {data[acc_no]['name']}
    1. Credit
    2. Debit
    3. CheckBalance
    4. Logout
    """
    print(menu)
    choice = int(input("Enter your choice : "))
    if choice == 1 :
        credit(acc_no)
    elif choice == 2 :
        debit(acc_no)
    elif choice == 3:
        chk_bal(acc_no)
    elif choice == 4 :
        print("\n sucessfully logout!")
        return 0
    else :
        print("\n\t\t...Invalid Choice...\n")
    print("\n\t.... Redirecting....\n")
    time.sleep(2)
    sub_menu(acc_no)
        
    
```

```python
import time
from getpass import getpass
def login():
    """This function for login service"""
    acc_no = input("Enter your account number: ").strip()
    if acc_no in data :
        password = getpass("Password : ")
        if data[acc_no]['password'] == password :
            #print("\n... Login is Successfull ...\n")
            time.sleep(2)
            return acc_no
        else :
            print("\n\t .... Invalid Password Try Again ...\n")
            time.sleep(2)
            return False
            
    else :
        print("\n\t Invalid User Account number\n")
        print("\n\t If you are New user pls Signup first")
        return False
```

```python
def signup():
    """This function for the Signup Service"""
    global last_account_num
    name = input("Enter your name : ").strip()
    password = getpass("Password : ")
    bal = eval(input("Enter initial amount : "))
    acc_no = str(last_account_num + 1)
    data[acc_no] = {'name':name,'bal':bal,'password':password}
    last_account_num+=1
    print("\n\n .... Creating Your Account .......\n\n ..... Please wait!..... \n\n")
    time.sleep(2)
    print(f"\n\n Write Down Your Account Number {acc_no} and use them to login.")
    time.sleep(2)
```

```python
# for admin 
def all_users():
    for key,value in data.items():
        print("Account number : ",key)
        for k,v in value.items():
            print(f"{k}={v}".rjust(100))
        print("\n")
```

```python

def main_menu():
    menu = f"""
    Today : {time.ctime()}
    1. Login
    2. SignUp
    3. Admin
    4. Exit
    """
    print(menu)
    choice = int(input("\nEnter your Choice : "))
    if choice == 1 :
        acc_no=login()
        if acc_no:
            sub_menu(acc_no)
            #print("\nLogin successfully done ")
        print("\n\n\t.....Redirecting to main menu....\n")
        time.sleep(5)
        main_menu()
    elif choice == 2 :
        signup()
        print("\nsuccessfully Sign up  Dude!")
        print("\n\n\t.....Redirecting to main menu....\n")
        time.sleep(5)
        main_menu()
    elif choice == 3 :
        admin()
        print("\n\n\t.....Redirecting to main menu....\n")
        main_menu()
    elif choice == 4 :
        print("\n\t\t\tBye Bye! Thank you , Visit Soon!\n")
        return 0
    else :
        print("\n\n Invalid Choice Try Again \n")
        main_menu()
```
